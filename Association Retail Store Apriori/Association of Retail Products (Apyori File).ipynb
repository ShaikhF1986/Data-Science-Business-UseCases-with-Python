### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv",header = None)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Conversion into the list

```python
transaction = []
for i in range(len(dataset)):
    transaction.append([str(dataset.values[i,j]) for j in range(0,len(dataset.iloc[0]))])
```

### Importing of the Apyori library for the association rules and classes

```python
from apyori import apriori
```

### Training of the dataset

```python
association_rules = apriori(transaction,min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

### Visualisation of the result

```python
association_results = list(association_rules)
```

```python
print(len(association_results)) 
```
